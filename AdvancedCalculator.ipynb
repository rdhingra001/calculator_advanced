This is the source code for an advanced calculator that I engineered, which is capable of computing the 4 operations, as well as square roots and exponents.
```
import math


# This is a class that was made to organize all my functions relating to the basic operations
class basicFunctions:

    # Function for calculating addition
    def addition():

        first_num = int(input("Please enter the first number that you would like to add. "))
        sec_num = int(input("Please enter the second number that you would like to add. "))
        total = first_num + sec_num
        print(first_num, " +", sec_num, " = ", total)

    # Function for calculating subtraction
    def subtraction():
        first_num = int(input("Please enter the first number that you would like to subtract. "))
        sec_num = int(input("Please enter the second number that you would like to subtract. "))
        difference = first_num - sec_num
        print(first_num, " -", sec_num, " = ", difference)
    
    # Function for calculating multiplication
    def multiplication():
        first_num = int(input("Please enter the first number that you would like to multiply. "))
        sec_num = int(input("Please enter the second number that you would like to mulitply"))
        product = first_num * sec_num
        print(first_num, " X", sec_num, " = ", product)

    # Function for calculating division
    def division():
        first_num = int(input("Please enter the first number that you would like to divide. "))
        sec_num = int(input("Please enter the second number that you would like to divide. "))
        quotient = first_num / sec_num
        print(first_num, " /", sec_num, " = ", quotient)
    
    # What is run if selected "basic"
    def basicOperations():
        selected_operation = input("Please enter one of the four operations in the given space. ")

        accepted_inputs = ["addition", "subtraction", "multiplication", "division"]
        
        while selected_operation not in accepted_inputs:
            selected_operation = input("Invalid input! Please enter one of the four operations in the given space. ")

        if selected_operation == "addition":
            basicFunctions.addition()
        elif selected_operation == "subtraction":
            basicFunctions.subtraction()
        elif selected_operation == "multiplication":
           basicFunctions.multiplication()
        elif selected_operation == "division":
            basicFunctions.division()

 
# This is a class that was made to organize all my functions relating to the advanced operations
class advancedFunctions:

    # Function for calculating exponents
    def exponential():
        first_num = int(input("Please enter the number that you would like to raise to the given power. "))
        sec_num = int(input("Pleae enter the number that you would to be raised to the first number inputted. "))
        exponentTotal = first_num ** sec_num
        print(first_num, " raised to the ", sec_num, " equals to ", exponentTotal, ".")
        
    # Function for calculating square roots
    def squareRoot():
        num = int(input("Please enter the number of that you would like to find the square root of. "))
        sqRoot = math.sqrt(num)
        print("The square root of ", num, " equals ", sqRoot, ".")

    # What is run if selected "advanced"
    def advancedFunctions():
      selected_operation = input("Please enter either 'exponent' or 'square root', respectively representing either raising a number to a certain power, or finding the square root of a number. ")
      accepted_outputs = ["exponent", "square root"]

      while selected_operation not in accepted_outputs:
        selected_operation = input("Invalid input! Please enter either 'exponent' or 'square root', respectively representing either raising a number to a certain power, or finding the square root of a number. ")
      
      if selected_operation == "exponent":
        advancedFunctions.exponential()
      elif selected_operation == "square root":
        advancedFunctions.squareRoot()
    
    

# First line of code that is run, determining the difficulty of the operations
def determineDifficulty():
    selected_difficulty = input("Please specify if you would like to carry out basic or advanced operations. Note that basic operations are your 4 operations, and advanced operations include exponents and square roots. ")
    if selected_difficulty == "basic":
        basicFunctions.basicOperations() 
    elif selected_difficulty == "advanced":
        advancedFunctions.advancedFunctions()
    else:
        selected_difficulty = input("Invalid input! Please specify if you would like to carry out basic or advanced operations. Note that basic operations are your 4 operations, and advanced operations include exponents and square roots. ")


determineDifficulty()
```
